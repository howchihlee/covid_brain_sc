```python
import os
import scipy.sparse as sp_sparse
import scanpy as sc
import numpy as np
from anndata import AnnData
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import harmonypy as hm
```

```python
folder = '../../processed_data/clean_data/'
df_meta = pd.read_csv(folder + 'metadata.csv', index_col = 0)
df_gene = pd.read_csv(folder + 'gene_expression_v10k.csv', index_col = 0)

```

```python
sampleid = ['-'.join([a[0], str(a[1])]) for a in zip(df_meta.pid, df_meta.tissue_id)]
```

```python
import scrublet
import scrublet as scr
scores = []
for k, df_sub in df_gene.groupby(sampleid): 
    scrub = scr.Scrublet(df_sub.values)
    doublet_scores, predicted_doublets = scrub.scrub_doublets()
    score = pd.DataFrame(doublet_scores, index = df_sub.index)
    scores.append(score)
doublet_scores = pd.concat(scores)
```

```python
doublet_scores.columns = ['doublet_score']
doublet_scores.to_csv('./seurat/doublet_scores_scrublet.csv')
```
