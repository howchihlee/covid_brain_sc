```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('./dream_test_toptfs.csv')
df['z.std'] = np.abs(df['z.std'])
df = df[df['overall_tissue_fraction'] > 0.1]
```

```python
set(df.Tissue)
```

```python
tmp = df.copy()
tmp = tmp.groupby(['Tissue', 'CellType']).mean().sort_values('z.std')
tmp['label'] = [f'{c}:{t}' for t, c in tmp.index]
```

```python
tmp1 = df.copy()
tmp1['label'] = [f'{c}:{t}' for t, c in zip(tmp1.Tissue, tmp1.CellType)]
tmp1.index = tmp1.label
```

```python
selected = tmp.iloc[-1:-10:-1, ].label.values
```

```python
tmp1 = tmp1.loc[selected]
```

```python
import matplotlib.pyplot as plt
sns.boxplot(data = tmp1, x = 'label', y = 'z.std')
sns.stripplot(data = tmp1, x = 'label', y = 'z.std', color = 'black')
plt.ylabel('absolute z score')
plt.xlabel('')
plt.xticks(rotation = 90)
plt.savefig('TF_zscores.pdf', bbox_inches='tight')
```
