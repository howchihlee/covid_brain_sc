```python
import pandas as pd
import glob
```

```python
res = []

goids = set()
gid2term = {}

for fn in glob.glob('./markers/*_go.csv'):
    df = pd.read_csv(fn, index_col = 0)
    df = df[df.Ont == 'BP']
    ct = fn.split('/')[-1].split('_')[0]
    goids.update(df.sort_values('P.DE').index[:3])
    res.append((ct, df)) 
goids = sorted(goids)
```

```python
goid2term = {g:t for t, g in zip(df.Term, df.index)}
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = -np.log10(np.vstack([df.loc[goids]['P.DE'].values for ct, df in res]))
df = pd.DataFrame(df)

df.columns = [goid2term[g] for g in goids]

df.index = [c for c, _ in res]
#fig, ax = plt.subplots(1, 1, figsize = (6, 15))
sns.clustermap(df.T, cmap = 'Greens', dendrogram_ratio=0.15, figsize = (12, 12))
plt.savefig("figs/go_enrichment_celltype.pdf")
plt.savefig("figs/png/go_enrichment_celltype.png", dpi = 600, bbox_inches='tight')
```
